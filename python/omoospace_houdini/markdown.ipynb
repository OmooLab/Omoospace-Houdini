````python
from pathlib import Path
import json

NODE_CONFIG_PATH = Path.cwd().parent.parent / "nodes.json"

with NODE_CONFIG_PATH.open('r') as f:
    data = json.load(f)
    EXPORT_PARMS = data['export_parms']

line_str_list = []
for node_type in EXPORT_PARMS.keys():
    parms = EXPORT_PARMS[node_type]
    parm_str_list = []
    for parm_name in parms.keys():
        parm_value = parms[parm_name]
        parm_name = parm_name.replace("_","\_")
        parm_str_list.append(f"{parm_name}: ``` {parm_value} ```")
        
    parms_str = " <br> ".join(parm_str_list)
    line_str = f"| {node_type} | {parms_str} |"
    line_str_list.append(line_str)

print("\n".join(line_str_list))
````
